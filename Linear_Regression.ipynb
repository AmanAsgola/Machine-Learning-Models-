```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.array([1,2,4,6,8])
y=np.array([3,4,8,10,15])
df = pd.DataFrame({'X': X, 'y': y})
X_mean=np.mean(X)
y_mean=np.mean(y)
```

```python
X=df['X'].astype(int).to_numpy().reshape(-1, 1)
y=df['y'].astype(int)
print(X)
# # when many columns are given iloc()
# X=df.iloc[:,0]
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
plt.figure(figsize=(8,6)) 
plt.scatter(X, y, color='blue', label='Data Points') 
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line') 
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Slope (m)",round(model.coef_[0],2))
print("intercept (b)",round(model.intercept_,2))
```
